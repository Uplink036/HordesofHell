```python
import os
import pathlib
import pandas as pd
from difflib import SequenceMatcher
```

```python
dir = pathlib.Path(os.path.abspath(''))
compass_df = pd.read_csv(dir / "D4 Infernal Compass - Runs (Season 10).csv")
compass_df.head(10)
```

```python
a1 = compass_df["Alternativ 1"].drop_duplicates()
a2 = compass_df["Alternativ 2"].drop_duplicates()
a3 = compass_df["Alternativ 3"].drop_duplicates()
all_events = (set(list(a1)+list(a2)+list(a3)))
all_events = set([event.strip() for event in all_events])
number_of_events = len(all_events)
print(f"{number_of_events=}")
print(all_events)
```

```python
def check_event_similarity(events: list, cutoff: float = 0.85):
    similar_events = []
    for i in range(0, len(events)):
        current_event = events[i]
        for event in events[i+1:]:
            similarity_score = SequenceMatcher(None, current_event, event).ratio()
            if similarity_score > cutoff:
                similar_events.append((current_event, event))
    return similar_events
```

```python
if (error := check_event_similarity(list(all_events))) != []:
    error_output = "Found potential duplicate datapoints, please fix\n" + \
                    "\n".join([x1+" ~ "+ x2 for x1, x2 in error])
    raise ValueError(error_output)
```
